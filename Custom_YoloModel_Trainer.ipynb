```python
import fiftyone as fo
import fiftyone.zoo as foz
import os

# Load COCO 2017 dataset with only the 'person' class for all splits
splits = ['train', 'validation', 'test']

for split in splits:
    print(f"\n--- Loading {split} split ---")
    dataset = foz.load_zoo_dataset(
        "coco-2017",
        split=split,
        label_types=["detections"],
        classes=["person"],
        max_samples=5000,
        dataset_name=f"coco_person_{split}"
    )


    export_dir = "./datasets/coco_person_yolo"
    dataset.export(
        export_dir=export_dir,
        dataset_type=fo.types.YOLOv5Dataset,
        label_field="ground_truth",
        split=split,
        classes=["person"]
    )

print("\n✅ Export complete! Data saved in ./datasets/coco_person_yolo/")
```

```python
#No need to run this if you extracted the archive directly from the link
import os, shutil

base = "/content/datasets/coco_person_yolo"

# Create YOLO-style folder structure
for split in ["train", "validation", "test"]:
    split_dir = os.path.join(base, split if split != "validation" else "val")
    img_src = os.path.join(base, "images", split)
    lbl_src = os.path.join(base, "labels", split)
    img_dst = os.path.join(split_dir, "images")
    lbl_dst = os.path.join(split_dir, "labels")

    os.makedirs(img_dst, exist_ok=True)
    os.makedirs(lbl_dst, exist_ok=True)

    if os.path.exists(img_src):
        for f in os.listdir(img_src):
            shutil.move(os.path.join(img_src, f), img_dst)
    if os.path.exists(lbl_src):
        for f in os.listdir(lbl_src):
            shutil.move(os.path.join(lbl_src, f), lbl_dst)

print("Reorganized dataset to YOLO structure")

# Clean up old empty folders
shutil.rmtree(os.path.join(base, "images"), ignore_errors=True)
shutil.rmtree(os.path.join(base, "labels"), ignore_errors=True)
```

```python
import os

# Make sure dataset folder exists
os.makedirs("datasets", exist_ok=True)

# Create YOLO dataset YAML
yaml_content = """\
train: /content/datasets/coco_person_yolo/train/images
val: /content/datasets/coco_person_yolo/val/images
test: /content/datasets/coco_person_yolo/test/images

names:
  0: person
"""

yaml_path = "datasets/coco_person.yaml"

with open(yaml_path, "w") as f:
    f.write(yaml_content)

print(f"Created dataset YAML at {yaml_path}")
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
results = model.train(
    data="datasets/coco_person.yaml",
    epochs=50,
    imgsz=480,
    batch=8,
    device=0,
    project="./training",
    name="yolo11n_person",
    seed=18,
    workers=2,
    plots=True
)


#Vslidate performance
metrics = model.val()
print("Validation complete. mAP:", metrics.box.map)

#Run quick test
test_images = [
    "https://cdn.antaranews.com/cache/1200x800/2023/10/13/Pengendara-Sepeda-Motor-Trotoar-060323-aaa-5.jpg",
    "https://img.harianjogja.com/posts/2022/11/14/1117643/jalur-pedestrian-malioboro.jpg",
    "https://assets.bwbx.io/images/users/iqjWHBFdfxIU/i.KTm08H6tuM/v1/1200x810.jpg"
]

predictions = model.predict(source=test_images, imgsz=480)
print("Predictions complete.")
```

```python
#Generate dataset for int8 callibration
import os
import random
import shutil
from glob import glob

# === CONFIG ===
train_dir = "/content/datasets/coco_person_yolo/train/images"
val_dir = "/content/datasets/coco_person_yolo/val/images"
calib_dir = "/content/calib_imgs"
os.makedirs(calib_dir, exist_ok=True)

num_train = 70  # number from train
num_val = 30    # number from val

# === COLLECT IMAGES ===
train_images = glob(os.path.join(train_dir, "*.jpg"))
val_images = glob(os.path.join(val_dir, "*.jpg"))

# ensure we don't exceed available images
train_sample = random.sample(train_images, min(num_train, len(train_images)))
val_sample = random.sample(val_images, min(num_val, len(val_images)))

# merge and shuffle final list
calib_images = train_sample + val_sample
random.shuffle(calib_images)

# === COPY SELECTED IMAGES TO calib_imgs/ ===
for img in calib_images:
    shutil.copy(img, calib_dir)

# === WRITE calib_list.txt WITH ABSOLUTE PATHS ===
calib_list_path = os.path.join(calib_dir, "calib_list.txt")
with open(calib_list_path, "w") as f:
    for img in sorted(glob(os.path.join(calib_dir, "*.jpg"))):
        f.write(os.path.abspath(img) + "\n")

print(f"Done! {len(calib_images)} images copied to: {calib_dir}")
print(f"Calibration list saved at: {calib_list_path}")
```
